```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv(PATH + 'train.csv').replace('male', 1).replace('female', 0)

train = pd.concat([train, pd.get_dummies(train.Embarked, prefix="Emb")], axis=1)

train = pd.concat([train, pd.get_dummies(train.Pclass, prefix="Pclass")], axis=1)
train['Parch'] = train.Parch > 1
train['SibSp'] = train.SibSp > 0
train = train.replace(True, 1).replace(False, 0)
```

```python
test = pd.read_csv(PATH + 'test.csv').replace('male', 1).replace('female', 0)
test = pd.concat([test, pd.get_dummies(test.Embarked, prefix="Emb")], axis=1)
test = pd.concat([test, pd.get_dummies(test.Pclass, prefix="Pclass")], axis=1)
test['Have_fam'] = test.Parch + test.SibSp > 0
test['Parch'] = test.Parch > 0
test['SibSp'] = test.SibSp > 1
test = test.replace(True, 1).replace(False, 0)
```

```python
sample_submission = pd.read_csv(PATH + 'sample_submission.csv', index_col='PassengerId')
```

```python
features = ['Pclass_1', 'Pclass_2', 'Pclass_3', 'Sex', 'Emb_C', 'Emb_Q', 'Emb_S', 'SibSp', 'Parch']
target = 'Survived'
```

```python
#  NAN  
#train = train.fillna(0)
#test = test.fillna(0)
#  NAN  

#  NaN    
train_numeric = train.select_dtypes(include=[float, int])
train[train_numeric.columns] = train_numeric.fillna(train_numeric.mean())

test_numeric = test.select_dtypes(include=[float, int])
test[test_numeric.columns] = test_numeric.fillna(train_numeric.mean())  #      train
```

```python
#          numpy
x_train = train[features].values
x_test = test[features].values
x_train
```

```python
# 
from sklearn.preprocessing import KBinsDiscretizer

kbd = KBinsDiscretizer(n_bins=[4,4], encode='onehot-dense') #strategy='uniform' strategy='kmeans' )
#  
train_transformed = kbd.fit_transform(train[['Age', 'Fare']])
test_transformed = kbd.fit_transform(test[['Age', 'Fare']])
#    
kbd.bin_edges_
```

```python
x_train[:5]
```

```python
#         numpy
y_train = train[target].values
```

```python
y_train[0:5]
```

```python
#  
mean = x_train.mean(axis=0)
#  
std = x_train.std(axis=0)
x_train = x_train - mean
x_train /= std
x_test = x_test - mean
x_test /= std
```

```python
x_train = np.column_stack( ( train_transformed, x_train ) )
x_test = np.column_stack( ( test_transformed, x_test ) )
```

```python
model = Sequential()
model.add(Dense(190, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
print(model.summary())
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train,
                    y_train,
                    epochs=20,
                    batch_size=120, # batch_size    32
                    validation_split=0.1,
                    verbose=2)
```

```python
plt.plot(history.history['accuracy'],
         label='     ')
plt.plot(history.history['val_accuracy'],
         label='     ')
plt.xlabel(' ')
plt.ylabel('  ')
plt.legend()
plt.show()
```

```python
predictions = model.predict(x_test)
```

```python
"""             """
#sample_submission[target] = predictions
sample_submission[target] = [0 if pred < 0.5 else 1 for pred in predictions]
```
